```python
#example 6
def plot_slope(X,Y):
  Xs = X[1:] - X[:-1]
  Ys = Y[1:] - Y[:-1]
  plt.plot(X[1:],Ys/Xs)

X = np.linspace(-3,3,100)
Y = np.exp(-X**2)

plt.plot(X,Y)
plot_slope(X,Y)

plt.show()
```

```python
#example 13
data = [[5,25,50,20],
        [4,23,51,17],
        [6,22,52,19]]

X = np.arange(4)

plt.bar(X+0,data[0],color ='b',width = 0.25)
plt.bar(X+0.25,data[1], color='g', width = 0.25)
plt.bar(X+0.5,data[2],color = 'r', width = 0.25)
plt.show()
```

```python
#example 14 Stacked bar

A = [5.0, 30.0, 45.0, 22.0]
B = [5.0, 25.0, 50.0, 20.0]

X =range(4)


plt.bar(X,A, color = 'b')
plt.bar(X,B, color = 'r', bottom = A)


plt.show()
```

```python
# example 15
A = np.array([5., 30., 45., 22.])
B = np.array([5., 25., 50., 20.])
C = np.array([1., 2., 1., 1.])
X = np.arange(4)

plt.bar(X,A,color = 'b')
plt.bar(X,B,color = 'g', bottom = A)
plt.bar(X,C,color = 'r', bottom = A + B)

plt.show()
```

```python
#example 16
women_pop = np.array([5.0,10.0,45.0,22.0])
mem_pop = np.array( [5., 25., 50., 20.])
X = np.arange(4)
plt.barh(X, women_pop, color = 'r')
plt.barh(X,  -mem_pop, color = 'b')

plt.show()
```
